```python
# Function to analyze post engagement
def analyze_post_engagement(likes_list):
    
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Display output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example data (likes received over time)
likes_data = [200, 150, 300, 250, 150]

# Function call
analyze_post_engagement(likes_data)
```

```python
# Function to monitor medicine stock
def check_medicine_stock(stock):

    # Conditional logic to check stock level
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    # Display output
    print("Medicine Stock:", stock)
    print("Status:", status)


# Example stock value
medicine_stock = 6

# Function call
check_medicine_stock(medicine_stock)
```

```python
# Function to analyze rainfall data
def rainfall_checker(rainfall_data, required_level):

    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)

    # Check rainfall adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Display output
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Example rainfall data (weekly rainfall values)
rainfall_values = [70, 75, 68, 80, 67]
required_rainfall = 70

# Function call
rainfall_checker(rainfall_values, required_rainfall)
```

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):

    # Convert list to set
    unique_users = set(usernames)

    # Check for duplicates
    if len(unique_users) < len(usernames):
        result = "Yes"
    else:
        result = "No"

    # Display output
    print("Duplicate Accounts Found:", result)


# Example usernames
user_list = ["hema01", "alex22", "john99", "hema01", "rita07"]

# Function call
detect_duplicate_accounts(user_list)
```

```python
# Function to evaluate application health
def application_health_checker(error_count):

    # Conditional checks
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    # Display output
    print("Error Count:", error_count)
    print("System Status:", status)


# Example error count
errors = 7

# Function call
application_health_checker(errors)
```

```python
# Function to validate daily transaction amount
def transaction_limit_checker(amount):

    daily_limit = 50000  # ₹50,000 limit

    # Conditional logic
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    # Display output
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example transaction amount
transaction_amount = 60000

# Function call
transaction_limit_checker(transaction_amount)
```

```python
# Function to calculate attendance and check eligibility
def attendance_checker(attendance_list):

    total_classes = len(attendance_list)
    attended_classes = 0

    # Loop to count attended classes
    for status in attendance_list:
        if status == 1:     # 1 means present
            attended_classes += 1

    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    # Display output
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", result)


# Example attendance data (1 = Present, 0 = Absent)
student_attendance = [1, 1, 0, 1, 1]

# Function call
attendance_checker(student_attendance)
```
