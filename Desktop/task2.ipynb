```python
sentence = input("Enter a sentence: ")

# Step 1: Split sentence into words
words = sentence.split()

# Step 2: Convert to set (removes duplicates)
unique_words = set(words)

# Step 3: Print results
print("Count of unique words:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Step 1: Store employee names and salaries
employees = {}

n = int(input("Enter number of employees: "))

for i in range(n):
    name = input("Enter employee name: ")
    salary = float(input("Enter salary: "))
    employees[name] = salary

# Step 2: Find highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Step 3: Display result
print("Employee with highest salary:", highest_employee)
print("Highest salary:", highest_salary)
```

```python
numbers = [45, 22, 89, 10, 66]

# Step 1: Initialize
maximum = numbers[0]
minimum = numbers[0]

# Step 2: Loop and compare
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Step 3: Print results
print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
text = "pythonpp"   # Example input

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```
