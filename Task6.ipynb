```python
# Function to manage smart parking system
def smart_parking_system(capacity, vehicle_logs):

    parked = 0
    peak_usage = 0

    # Process vehicle logs
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak parking usage
        if parked > peak_usage:
            peak_usage = parked

    # Check parking status
    if parked > capacity:
        status = "Alert: Parking Exceeded Capacity"
    else:
        status = "Available"

    # Display output
    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)


# Sample Input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Function call
smart_parking_system(parking_capacity, vehicle_logs)
```

```python
# Function to estimate delivery time
def delivery_time_estimator(distance, traffic, weather):

    # Base delivery time (5 minutes per km)
    eta = distance * 5

    # Traffic condition adjustment
    if traffic == "High":
        eta += 10
    elif traffic == "Medium":
        eta += 5
    # Low traffic → no extra delay

    # Weather condition adjustment
    if weather == "Rainy":
        eta += 5
    elif weather == "Stormy":
        eta += 15
    # Clear weather → no delay

    # Display output
    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
distance_km = 8
traffic_level = "High"
weather_condition = "Rainy"

# Function call
delivery_time_estimator(distance_km, traffic_level, weather_condition)
```

```python
# Function to classify server load
def server_load_classifier(cpu_readings):

    total = 0

    # Loop to calculate total CPU usage
    for reading in cpu_readings:
        total += reading

    # Calculate average CPU load
    average_cpu = total / len(cpu_readings)

    # Classify server status
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # Display output
    print("Average CPU Load:", int(average_cpu), "%")
    print("Server Status:", status)


# Sample Input
cpu_data = [45, 60, 70, 85, 90]

# Function call
server_load_classifier(cpu_data)
```

```python
# Function to monitor classroom resource usage
def resource_usage_monitor(resource_usage):

    overused = []

    # Assume overuse threshold = 8 hours
    threshold = 8

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    # Determine energy alert
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    # Display output
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function call
resource_usage_monitor(usage_data)
```

```python
# Function to manage event registrations
def event_registration_controller(capacity, registrations):

    confirmed = 0

    # Loop to process registrations
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1

    # Calculate waitlisted users
    waitlisted = registrations - confirmed

    # Registration status
    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"

    # Display output
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_capacity = 100
total_registrations = 105

# Function call
event_registration_controller(event_capacity, total_registrations)
```
