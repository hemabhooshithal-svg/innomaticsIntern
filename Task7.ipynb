```python
class SmartLight:

    def __init__(self, name):
        self.light_name = name
        self.is_on = False   # initially OFF

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        if self.is_on:
            print(self.light_name + " is ON")
        else:
            print(self.light_name + " is OFF")


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()

# Output
light.display_status()
```

```python
class Employee:

    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
emp = Employee("Rahul", "EMP102", "IT")

# Output
emp.display_id_card()
```

```python
class Contact:

    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact details
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
contact1 = Contact("Anita", "9876543210")

# Output
contact1.display_contact()
```

```python
class Product:

    # Constructor
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Method to print price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Input
product1 = Product("Headphones", 2499)

# Output
product1.display_price_tag()
```

```python
class Movie:

    # Constructor
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Input
movie1 = Movie("Inception", 4.8)

# Output
movie1.display_movie()
```

```python
class Delivery:

    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Input
delivery1 = Delivery("Suman", "Hyderabad")

# Output
delivery1.display_delivery_details()
```
